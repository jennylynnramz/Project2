```python
import pandas as pd
```

```python
redfin_original_csv = pd.read_csv('redfin_2020-06-26-14-22-58.csv')
redfin_original_csv.head()
```

```python
# list(redfin_original_csv.columns)
```

```python
redfin_columns_removed = redfin_original_csv[['PROPERTY TYPE', 'ADDRESS', 'CITY', 'STATE OR PROVINCE', 'ZIP OR POSTAL CODE', 'PRICE', 'BEDS', 'BATHS', 'LOCATION', 'SQUARE FEET', 'YEAR BUILT', '$/SQUARE FEET', 'HOA/MONTH', 'LATITUDE', 'LONGITUDE']]
redfin_columns_removed.head()
```

```python
redfin_columns_removed["COORDINATES"] = redfin_columns_removed["LATITUDE"].astype(str) +"," + redfin_columns_removed["LONGITUDE"].astype(str)
redfin_columns_removed.head()
```

```python
redfin_location_combined = pd.DataFrame(redfin_columns_removed)
```

```python
redfin_location_combined
```

```python
redfin_location_combined['YEAR BUILT'] = redfin_location_combined['YEAR BUILT'].astype(str)
```

```python
redfin_location_combined['BUILT YEAR'] = redfin_location_combined['YEAR BUILT'].astype(str).str[:4]
redfin_location_combined
```

```python
redfin_location_combined.head()
```

```python
redfin_year_new = redfin_location_combined[['PROPERTY TYPE', 'ADDRESS', 'CITY', 'STATE OR PROVINCE', 'ZIP OR POSTAL CODE', 'PRICE', 'BEDS', 'BATHS', 'LOCATION', 'SQUARE FEET', '$/SQUARE FEET', 'HOA/MONTH', 'LATITUDE', 'LONGITUDE', 'BUILT YEAR', 'COORDINATES']]
```

```python
redfin_year_new.head()
```

```python
redfin_year_new.to_csv('../Resources/clean_redfin_data.csv', index = True)
```
